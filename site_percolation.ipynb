```python
import numpy as np
import matplotlib.pyplot as plt

def site_percolation(n, p):
    # Generate a random array with probabilities of each site being occupied
    lattice = np.random.rand(n, n) < p
    return lattice.astype(int)

def plot_percolation(lattice):
    plt.imshow(lattice, cmap='binary', interpolation='nearest')
    plt.title('Site Percolation for p = 0.55')
    plt.savefig('site_percolation.pdf')
    plt.show()

# Set lattice size and probability of occupation
n = 200
p = 0.55

# Generate site percolation lattice
lattice = site_percolation(n, p)

# Plot the lattice
plot_percolation(lattice)
```
